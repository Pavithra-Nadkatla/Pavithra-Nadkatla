```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
data=pd.read_csv('Titanic-Dataset.csv')
```

```python
#1.	What is the distribution of passengers' ages on the Titanic?
sns.histplot(data['Age'],bins=10,kde='True')

plt.xlabel=("age")
plt.ylabel=('Count')
plt.set_title=("Distribution of passengers'age")
plt.show()
```

```python
#2.	How many passengers survived vs. not survived? Visualize the count using a bar plot
d=data['Survived'].value_counts()
d.index=['not_survived','survived']
d
x = d.index.tolist()
y = d.values.tolist()
plt.bar(x,y,data=data['Survived'])
plt.ylabel=('Count')
plt.title=("Passenger's survival count")

#sns.barplot(x=d.index,y=d.values,data=data['Survived'])
#plt.bar(x,y,data=data['Survived'])
#sns.countplot(x='Survived')
```

```python
#3.	What is the distribution of passenger class (Pclass) on the Titanic?

sns.countplot( x='Pclass',data=data)
plt.xlabel=('Passenger class type')
plt.ylabel=('Count')

plt.xticks(ticks=[0,1,2], labels=['p1','p2','p3']);

```

```python
#4.	Visualize the distribution of passenger ages for each passenger class using a box plot

sns.boxplot(data,x='Pclass',y='Age')
plt.title=('Distribution of passenger ages wrt class')
plt.xticks(ticks=[0,1,2], labels=['p1','p2','p3']);
```

```python
#5.	Create a pair plot to explore relationships between numerical variables like age, fare, and the number of siblings/spouses (sibsp) aboard

d=data[['Age','Fare','SibSp']]
d
sns.pairplot(d,hue='SibSp')
```

```python
#6.	How does the survival rate differ based on passenger class? Visualize it using a bar plot.

sns.barplot(x='Pclass',y='Survived',data=data)
plt.title=('Survival rate wrt passenger class')
plt.xticks(ticks=[0,1,2], labels=['P1','P2','P3']);
```

```python
#19.	Create a line plot to visualize the changes in the average fare paid over different embarkation ports.
s_data = data[data['Embarked'] == 'S']
c_data = data[data['Embarked'] == 'C']
q_data = data[data['Embarked'] == 'Q']
sns.lineplot(x=s_data.index, y='Fare', data=s_data, estimator='mean',label='S')
sns.lineplot(x=c_data.index, y='Fare', data=c_data, estimator='mean',label='C')
sns.lineplot(x=q_data.index, y='Fare', data=q_data, estimator='mean',label='Q')
plt.set_title=('Average Fare by Embarkation Port')
plt.xlabel=('Embarkation Port')
plt.ylabel=('Average Fare')
plt.legend()
plt.show()
```
